```python
# Problem 4 code here 
lm2 = smf.ols(formula = "Perc_Economically_Disadvantaged ~ Perc_English_Language_Learner + Average_Class_Size + Average_Salary ", data = Sch).fit()
lm2.summary()
```

```python
## After adding the rest two predictors, the total R-squre increased from .37 to .49. The two prectors were significantly predictive of Perc_Economically_Disadvantaged, 
## specifically, both Average_Class_Size and Average_Salary were negatively associated with Perc_Economically_Disadvantaged. 
## The intercept in the updated model changed from 21.96 to 95.43. The regression coefficient of Perc_English_Language_Learner changed from 1.03 to 1.22. 
```

```python
lm3 = smf.ols(formula = "Perc_Economically_Disadvantaged ~ Perc_English_Language_Learner + Average_Class_Size + Average_Salary + School_Type", data = Sch).fit()
lm3.summary()

## After adding School_Type to the model, the effect size and model fit did not change from the model lm2. The regression coefficient of the three predictors used in model lm2 did not change.
## However, the coefficient of the intercept changed and the coefficient of the School_Type variable had the same coefficient as the intercept. 
## It might because the School_Type variable was a binary variable and there was significant differences in Perc_Economically_Disadvantaged between public school and charter schools. 
```
